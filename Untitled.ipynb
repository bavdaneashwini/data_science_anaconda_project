```python
#Create a bar plot of the total  number of homeless people by region in the 'homelessness' Dataframe . Order the bar in descending order. Bonus: create a horizantal bar chart.
```

```python
df_home = pd.read_csv('https://raw.githubusercontent.com/julito14/Data-Manipulation-w-Pandas-Project-1/main/homelessness.csv')
```

```python
df_home.head(10)
```

```python
df_home.info()
```

```python
homeless_per_region = df_home.groupby('region').agg({'individuals':'sum'}).sort_values('individuals',ascending=True)
```

```python
homeless_per_region.head(10)
```

```python
plt.barh(homeless_per_region.index,homeless_per_region.individuals,color='#ff006e')

plt.show()
```

```python
#Create a line plot with two lines representing the temperatures in Toronto and Rome(create this dataframe from the temperatures.csv file) make sure to properly label your plot , Bonus : add a legend for the two lines
```

```python
df_temp = pd.read_csv('https://raw.githubusercontent.com/julito14/Data-Manipulation-w-Pandas-Project-1/main/temperatures.csv')
```

```python
df_temp.head(5)
```

```python
temp_rome = df_temp[df_temp.city=='Rome']
temp_rome.head(5)
```

```python
temp_toronto= df_temp[df_temp.city=='Toronto']
temp_toronto.head(5)
```

```python
temp_toronto.date = pd.to_datetime(temp_toronto.date)
```

```python
type(temp_toronto.date[15840])
```

```python
temp_rome.date = pd.to_datetime(temp_rome.date)
```

```python
plt.figure(figsize=(20,8))
plt.plot(temp_rome.date,temp_rome.avg_temp_c,label='Rome')
plt.plot(temp_toronto.date,temp_toronto.avg_temp_c,label='Toronto')
plt.legend(loc='best',framealpha=0.2)
plt.ylabel("Avg. Temperature")
plt.show()
```
